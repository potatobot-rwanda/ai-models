```python
lines = open("test.txt").readlines()
lines = list(filter(lambda x:len(x)>10, lines))
lines = [line[0:-1] for line in lines]
headings = lines[0]
lines = list(filter(lambda x:x != headings, lines))
lines = [line.split(" ") for line in lines]

headings = ['DISTRICT NAME',
 'SECTOR NAME',
 'CELL NAME',
 'VILLAGE NAME',
 'VILLAGE CODE',
 'RATE YEAR',
 'AGRICULTURE',
 'COMMERCIAL',
 'ECONOMIC',
 'FOREST',
 'INDUSTRY',
 'LIVERSTOCK',
 'RESIDENTIAL',
 'SCIENTIFIC',
 'SOCIAL',
 'TOURISM']
df = pd.DataFrame(lines[1:50], columns=headings)
df
```

```python
with gzip.open("lgt1.txt.gz", "r") as f:
    lines = f.readlines()
    lines = [line.decode("utf-8")[0:-1] for line in lines]

print("\n".join(lines[0:100]))
```

```python
from typing import defaultdict

df = defaultdict(list)

FINISHED_COLUMN = 0
READ_CELLS = 1

status = FINISHED_COLUMN

current_column = None
for line in lines:
    if line == "":
        status = FINISHED_COLUMN

    elif status == FINISHED_COLUMN:
        current_column = READ_CELLS
        current_column = line
        status = READ_CELLS

    elif status == READ_CELLS:
        df[current_column].append(line)

for key, value in df.items():
    print(key, len(value))
```
